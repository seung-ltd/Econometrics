```python
df2 = pd.read_stata("state_socioeconomic_1950.dta")
df2.head()
```

```python
df3 = pd.merge(df, df2, on='statenam')
df3.head()
```

## (d)

```python
for row in df3:
    df3['pct_urb950'] = df3['urb950']/df3['totpop']*100
    df3['pct_mtot'] = df3['mtot']/df3['totpop']*100
    df3['pct_black_pop'] = df3['black_pop']/df3['totpop']*100
    df3['pct_col_edu'] = df3['col_edu']/df3['totpop']*100
df3.head()
```

## (e)

```python
import matplotlib.pyplot as plt
from adjustText import adjust_text
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df3['pct_urb950'], df3['pct_tvhh'])

texts = []
for i, txt in enumerate(df3['state_abbrev']):
    texts.append(plt.text(df3['pct_urb950'][i], df3['pct_tvhh'][i], txt))
adjust_text(texts)

plt.title('Relationship between % TV Households & % Urban Population')
plt.xlabel('% Urban Population (1995)')
plt.ylabel('% TV Households')
plt.grid()
```

## (f)

```python
import statsmodels.api as sm
```

```python
Y = df3['circ_per1000']
X = df3['pct_tvhh']
X1 = sm.add_constant(X)
model = sm.OLS(Y, X1).fit()
print(model.summary())
```

### Answer:
The regression coefficient on pct_tvhh can be interpreted as follows: for each percentage point increase in the percentage of tv households there would be an expected average change in the number of daily newspaper circulations per 1000 population by 7.24 percentage points. This is assuming that there are no other variables that could affect this relationship, given there are no controls.

## (g)

```python
Y = df3['circ_per1000']
X = df3[['pct_tvhh', 'pct_urb950']]
X1 = sm.add_constant(X)
model = sm.OLS(Y, X1).fit()
print(model.summary())
```

### Answer:
The regression coefficient on pct_tvhh is now -1.41 instead of 7.24 indicating an inverse relationship when accounting for the urban population percentage variable. Moreover, the relationship is not statistically significant since the P value is very high at 0.541. This tells us that the relationship between newspaper circulation and % TV households is not causal and just an example of omitted variable bias causing upward bias in the previous regression.

## (h)

```python
coefficients = model.params
std_errors = model.bse
conf_int = model.conf_int()

if 'const' in coefficients:
    coefficients = coefficients.drop('const')
    conf_int = conf_int.drop('const')

fig, ax = plt.subplots()
coefficients.plot(kind='bar', ax=ax, color='blue', yerr=std_errors[coefficients.index] * 1.96)
ax.set_ylabel('Coefficient Value')
ax.set_xlabel('Variables')
ax.set_title('Regression Coefficients & Confidence Intervals')

for i, coef in enumerate(coefficients):
    ax.errorbar(i, coef, yerr=[[coef - conf_int[0][i]], [conf_int[1][i] - coef]], fmt='o', color='red', capsize=5)

plt.xticks(rotation=0)
```

```python
df4 = pd.read_stata("election_state.dta")
df4.head()
```

```python
df5 = pd.merge(df3, df4, on='statenam')
df5.head()
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols('rep52 ~ theater_per1000 + circ_per1000 + pct_tvhh + pct_urb950 + pct_mtot + pct_black_pop + pct_col_edu + C(region)', data=df5).fit(cov_type='HC1')
model.summary()
```

```python
results_df = pd.DataFrame({
    'Parameter': model.params.index,
    'Coefficient': model.params.values,
    'Std. Error': model.bse.values,
    't-values': model.tvalues.values,
    'P>|t|': model.pvalues.values,
    '[0.025': model.conf_int().iloc[:, 0],
    '0.975]': model.conf_int().iloc[:, 1]
})

results_df.to_excel("regression_results.xlsx", index=False)
```
